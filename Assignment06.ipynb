```python
####### assignment06.py #######

import numpy as np
import matplotlib.pyplot as plt

num     = 201
std     = 20
a       = 2 
b       = 10

n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-100,100,num)
y1      = a * x + nn * std + b 
y2      = a * x + b

plt.plot(x, y1, 'b.', x, y2, 'k--')
plt.show()

# x  : x-coordinate data
# y1 : (noisy) y-coordinate data
# y2 : (clean) y-coordinate data 
# y = f(x) = a * x + b
```

Plot the noise data

```python
### noise data plots ###
plt.subplot(1,2,1)
plt.plot(x, y1, 'b.')

plt.subplot(1,2,2)
plt.plot(x, y1)

plt.show()
```

Plot the clean data

```python
#### clean data plot #####
plt.plot(x, y2)
plt.show()
```

Calculate m, c in m*x + c with noise data set

```python
### examing coeffient ###

A = np.vstack([x, np.ones(len(x))]).T

###calculate m, c using np.linalg.lstsq

m, c = np.linalg.lstsq(A, y1)[0]

### print (a,b), (m,c), error ###
print(a, b)
print(m, c)
print("Error:", a - m, b - c)
```

Plot the line that fits the noisy data by the least square error

```python
### make fit line with m, c ###

yy = m*x + c
plt.plot(x, yy, x, y1, 'k--')
plt.show()
```
